##  Flatten the dataset

```python
X_train_F = X_train.reshape(len(X_train),28*28)
X_test_F = X_test.reshape(len(X_test),28*28)
```

# Model

## ANN

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation= 'sigmoid')
])

#Dense mean that all the neuron in the first layer are connected to all the other in the 2nd layer
#784 are the Xi
#10 is the output
#sigmoid is the activation function
```

```python
model.compile(optimizer ='adam',
              loss = "sparse_categorical_crossentropy",
              metrics = ['accuracy'])
model.fit(X_train_F,y_train,epochs=7)
              
# Optimizer Is used for better precision 
# Loss is the loss function 
# metrics is the objective of the optimization, in our case, to make it more accurate 
# epoches is the number of iteration
```

89% Accruacy

# Note

If you have poor precision, you need to scale your values

- X_train = X_train/255
- X_test = X_test/255

Your array, the value would be between 0 and 1

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_F = X_train.reshape(len(X_train),28*28)
X_test_F = X_test.reshape(len(X_test),28*28)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation= 'sigmoid')
])

model.compile(optimizer ='adam',
              loss = "sparse_categorical_crossentropy",
              metrics = ['accuracy'])
model.fit(X_train_F,y_train,epochs=7)
```

92.8% better !

```python
model.evaluate(X_test_F,y_test)
```

# Confusion Matrix in TF

```python
#convert prediction into a list to match with y_test
prediction_c = [np.argmax(i) for i in prediction]
# We will find the max probability in each value in prediction and stored into a list
```

```python
cm = tf.math.confusion_matrix(labels = y_test,predictions = prediction_c)
```

```python
import seaborn as sns
```

```python
sns.heatmap(
    cm,
    cmap="flare",
    annot=True,
    fmt='d',   
)

# The seaborn heatmap fmt help to show annot with different formatting.
```

# Hidden Layer

```python
model = keras.Sequential([
    keras.layers.Dense(50, input_shape=(784,),activation= 'relu'),
    keras.layers.Dense(10,activation= 'sigmoid')
])

model.compile(optimizer ='adam',
              loss = "sparse_categorical_crossentropy",
              metrics = ['accuracy'])
model.fit(X_train_F,y_train,epochs=7)

#The hidden layer (2row) does not need input_shape
#We will change the 10 value because we will be not connected to the output but to the hidden layer.
#50 isthe number of hidden layer wich is here less thant the inputs 784
#We changed the activation fucntion to relu
# You can add more hidden layer
```

98.2% ! Better 

```python
#Evaluate now the test set not the training
model.evaluate(X_test_F,y_test)
```

97% ! from 92.5%

```python
prediction = model.predict(X_test_F)
prediction_c = [np.argmax(i) for i in prediction]

cm = tf.math.confusion_matrix(labels = y_test,predictions = prediction_c)

sns.heatmap(
    cm,
    cmap="flare",
    annot=True,
    fmt='d',   
)
```

The number of errors decreased

# Without Flatten Arrays

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(50,activation= 'relu'),
    keras.layers.Dense(10,activation= 'sigmoid')
])

tb_callback = tf.keras.callbacks.TensorBoard(log_dir="logs/", histogram_freq=1)

#Log_dir is the directory where we save all the logs

model.compile(optimizer ='adam',
              loss = "sparse_categorical_crossentropy",
              metrics = ['accuracy'])
model.fit(X_train,y_train,epochs=7,callbacks=[tb_callback]) #We add it the the fitfuction, to supply all the information to tb callback

```

```python
#Saving the model
model.save("Quant")
```

# Post Training Quantization

```python
converter = tf.lite.TFLiteConverter.from_saved_model("Quant")
tflite_model = converter.convert()
```

```python
#It's the first approch tf model to tf lite model
```

```python
len(tflite_model)
#Now it's about 160 KB
```

```python
#Now let's use quantization
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model("Quant")
#We only need 1 line
converter.optimizations= [tf.lite.Optimize.DEFAULT]
#You have different mode of optimization, I choose default
tflite_quant_model = converter.convert()
```

```python
len(tflite_quant_model)
```

```python
#Saving the 2 models

with open("tflite_model.tflite", "wb") as f:
    f.write(tflite_model)
#.tflite is the extension for the model
#since it's byte data we will use write and byte mode "wb"

with open("tflite_quant_model.tflite", "wb") as f:
    f.write(tflite_quant_model)
```

```python
#You can see the size difference
```

# Quantization aware training

```python
import tensorflow_model_optimization as tfmot
#You need to install tensorflow model optimization
#pip install --user --upgrade tensorflow-model-optimization
```

```python
#1st step : create a q_model from the TF model
Q_model = tfmot.quantization.keras.quantize_model
#We will quantize the regular tensorflow model
Q_aware_model = Q_model(model)
```

```python
#2nd step Training the Q_model
```

```python
#Compile the model
Q_aware_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

Q_aware_model.summary()
```

```python
#train the model for 2 epochs
Q_aware_model.fit(X_train, y_train, epochs=2)
```

```python
Q_aware_model.evaluate(X_test, y_test)
#97% on test dataset !
```

```python
#Convert
converter = tf.lite.TFLiteConverter.from_keras_model(Q_aware_model)
#Quantization of conversion
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_Qaware_model = converter.convert()
```

```python
len(tflite_Qaware_model)
```

```python
#let's save it
with open("tflite_Qaware_model.tflite", 'wb') as f:
    f.write(tflite_Qaware_model)
```
